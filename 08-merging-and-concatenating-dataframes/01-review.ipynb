### 1. Configuration

```python
import pandas as pd
import numpy as np
```

```python
df_movies = pd.read_csv("IMDb movies.csv", low_memory=False)

movie_columns = [
    "imdb_title_id", 
    "title", 
    "year", 
    "genre", 
    "country"
    ]

df_movies = df_movies[movie_columns]
df_movies.head()
```

```python
df_ratings = pd.read_csv("IMDb ratings.csv", low_memory=False)

rating_columns = [
    "imdb_title_id", 
    "total_votes", 
    "mean_vote"
    ]

df_ratings = df_ratings[rating_columns]
df_ratings.head()
```

### 2.  Concatenate

#### 2.1. Concatenate -> Vertically (axis=0)

pd.concat([df1, df2])

```python
df1 = pd.DataFrame({
    "id": ["A", "B", "C", "D"],
    "age": [30, 23, 25, 22]
})

df2 = pd.DataFrame({
    "id": ["E", "F", "G", "H"],
    "age": [40, 21, 19, 24]
})
```

```python
pd.concat([df1, df2], axis=0, ignore_index=True)
```
