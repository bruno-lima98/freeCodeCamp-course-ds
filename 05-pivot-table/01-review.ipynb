### 3. pivot_table()

```python
df_sales = pd.read_excel("supermarket_sales.xlsx")
df_sales.head()
```

```python
columns = ["Quantity", "Rating", "Total", "Unit price", "cogs", "gross income", "gross margin percentage"]

df_sales.pivot_table(index="Gender", aggfunc="sum", values=columns)
```

```python
columns = ["Quantity", "Total"]

df_sales.pivot_table(
    index="Gender", 
    aggfunc="sum", 
    values=columns
    )
```

```python
columns = ["Total"]
categories = ["Product line"]

df_sales.pivot_table(
    index="Gender", 
    aggfunc="sum", 
    values=columns,
    columns=categories
    )
```
