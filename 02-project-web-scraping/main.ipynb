### 3. Reading multiple URLs

```python
# Link: root + season + league

"https://football-data.co.uk/mmz4281/" + "2526" + "/" + "E0" + ".csv"
```

```python
root = "https://football-data.co.uk/mmz4281/"
```

### 3.1. Multiple leagues

```python
leagues = ["E0", "E1", "E2", "E3", "EC"] # from site
dfs = []

for league in leagues:
    df = pd.read_csv(root + "2526" + "/" + league + ".csv")
    dfs.append(df)
```

```python
len(dfs)
```

```python
dfs[4].head()
```

### 3.2. Multiple seasons

```python
leagues = ["E0", "E1", "E2"] # from site
dfs = []

for league in leagues:
    for season in range(20,26):
        df = pd.read_csv(root + str(season) + str(season+1) + "/" + league + ".csv")
        df["season"] = str(season)

        dfs.append(df)
```

### 4. Organize Data in a Dictionary

```python
dict_countries = {
    "Spanish La Liga": "SP1",
    "Spanish Second Division": "SP2",
    "German Bundesliga": "D1",
    "English Premier League": "E0",
    "English League 1": "E1",
    "English League 2": "E2"
}
```

```python
dict_countries["Spanish La Liga"]
```

```python
dict_historical_data = {}

for league in dict_countries:
    dfs = []
    
    for season in range(20,26):
        df = pd.read_csv(root + str(season) + str(season+1) + "/" + dict_countries[league] + ".csv")
        df["season"] = str(season)

        dfs.append(df)

    df_concat = pd.concat(dfs)

    dict_historical_data[league] = df_concat
```
