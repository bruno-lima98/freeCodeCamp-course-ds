### 6.2. Adding a new column with an array

```python
language_score = np.random.randint(0, 101, size=1000)
len(language_score)
```

```python
df_exams["language score"] = language_score.round(2)
df_exams.head()
```

### 6.3. Adding a new column with assign()

When use assing()

- Add multiple columns in a single line of code
- When you need to overwrite the values on an existing column (best practice)

```python
score1 = np.random.randint(1, 101, size=1000)
score2 = np.random.randint(1, 101, size=1000)
```

```python
series1 = pd.Series(score1, index=np.arange(0, 1000))
series2 = pd.Series(score2, index=np.arange(0, 1000))
```

```python
type(score1)
```

```python
type(series1)
```

```python
df_exams = df_exams.assign(score_A=series1, score_B=series2)
```

```python
df_exams.head()
```

### 6.4. Adding a new column with intert()

```python
df_exams.insert(1, "test", series1)
df_exams.head()
```
