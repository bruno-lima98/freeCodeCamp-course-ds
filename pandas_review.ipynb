## 1. Basics

### 1.1 General

```python
import pandas as pd
import numpy as np
```

```python
data = np.array([[1, 0, 0], [0, 1, 0], [0, 0, 1]])
```

```python
columns = ["i", "j", "k"]
indexes = ["a", "b", "c"]

df = pd.DataFrame(data, columns=columns, index=indexes)
```

```python
df
```

```python
df = [
    [1, 2, 3], 
    [4, 5, 6], 
    [7, 8, 6]
    ]

columns = ["A", "B", "C"]

pd.DataFrame(df, columns=columns)
```

### 1.2 DataFrame from a dictionary

```python
states = ["São Paulo", "Paraná", "Maranhão", "Mato Grosso"]
population = [46649132, 11890517, 7018211, 3836399]
```

```python
dict_states = {"States":states, "Population":population}
```

```python
df_pop = pd.DataFrame(dict_states)
```

```python
df_pop
```
