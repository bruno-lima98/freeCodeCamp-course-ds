### 3. Add new column

```python
rating = []

for i in range(0,len(df)):
    rating.append(random.uniform(1,5))
```

```python
df["Critic Rating"] = rating
```

```python
df.head()
```

### 4. Rearanging the DataFrame

```python
df.columns
```

```python
new_columns = ["Name", "Author", "User Rating", "Critic Rating", "Reviews", "Price", "Year", "Genre"]
df = df[new_columns]
```

```python
df.head()
```

### 5. Operations

```python
df["Average Rating"] = (df["User Rating"] + df["Critic Rating"])/2
```

```python
df = df.round(1)
```

```python
df.head()
```

```python
df["Genre"].value_counts(normalize=True).round(2)
```

```python
rename = {
    "User Rating": "UR",
    "Critic Rating": "CR",
    "Average Rating": "AR"
}

df = df.rename(columns=rename)
```

```python
new_order = ["Name", "Author", "Price", "Year", "Genre", "Reviews", "UR", "CR", "AR"]
df = df[new_order]
```

```python
df.head()
```

### 6. Sorting

```python
df = df.sort_values(by=["UR", "CR"], ascending=False)
df
```
