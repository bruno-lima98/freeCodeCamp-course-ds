### 3. Create a column based on 1 condition

```python
base_price = 2000

np.where(df_laptops["Price_euros"] > base_price, "Expensive", "Cheap")
```

```python
df_laptops["Price_tier"] = np.where(df_laptops["Price_euros"] > base_price, "Expensive", "Cheap")
```

```python
df_laptops.head()
```

```python
df_laptops.value_counts("Price_tier")
```

### 5. Filter dataframe bases on a condition

```python
df_laptops[df_laptops["Company"] == "Apple"].head()
```

```python
## AND operator = &

# Laptops of company Apple costing more than 2.000 euros
filter = (df_laptops["Company"] == "Apple") & (df_laptops["Price_euros"] > 2000)

df_1 = df_laptops[filter]
```

```python
df_1.value_counts("Company")
```

```python
## OR operator = |

# Laptops of company Apple or HP
filter = (df_laptops["Company"] == "Apple") | (df_laptops["Company"] == "HP")

df_2 = df_laptops[filter]
```

```python
df_2.value_counts("Company")
```

```python
## OR and AND operators -> combine them using ()

# Laptops of company Apple or HP, costing more than 2.000 euros
filter = ((df_laptops["Company"] == "Apple") | (df_laptops["Company"] == "HP")) & (df_laptops["Price_euros"] > 2000)

df_3 = df_laptops[filter]
```

```python
df_3.value_counts("Company")
```

```python
df_3["Price_euros"].min()
```

### 6. Creating a conditional columns from more than 2 choices

```python
conditions = [
    (df_laptops["Price_euros"] > 3000),
    (df_laptops["Price_euros"] > 2000) & (df_laptops["Price_euros"] <= 3000),
    (df_laptops["Price_euros"] > 1000) & (df_laptops["Price_euros"] <= 2000),
    (df_laptops["Price_euros"] <= 1000),
]

values = ["Too Expensive", "Expensive", "Affordable", "Cheap"]
```

```python
np.select(conditions, values)
```

```python
df_laptops["Price_range"] = np.select(conditions, values)
```

```python
df_laptops.head()
```

```python
df_laptops["Price_range"].value_counts()
```

### 8. isin(): Multiple Filtering

```python
types = ["Ultrabook", "Notebook"]
companies = ["Apple", "HP"]

filter_1 = df_laptops["TypeName"].isin(types)
filter_2 = df_laptops["Company"].isin(companies)
```

```python
df_laptops[filter_1 & filter_2].head()
```

```python
df_laptops[filter_1 & filter_2]["Company"].value_counts()
```

```python
df_laptops[filter_1 & filter_2]["TypeName"].value_counts()
```

### 9.1. duplicated(): keep

### 10. drop_duplicated()

```python
df_laptops = pd.read_csv("laptop_price.csv", encoding="latin-1") # reseting df
```

```python
df_laptops.drop_duplicates(["Company"])[["Company", "Price_euros"]].value_counts()
```

```python
df_laptops.drop_duplicates(["Company"], keep="first")[["Company", "Price_euros"]]
```

```python
df_laptops.drop_duplicates(["Company"], keep="last", inplace=True, ignore_index=True)
```
