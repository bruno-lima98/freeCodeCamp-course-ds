### 4. Exercise

```python
conditions = [
    (df_laptops["Inches"] > 16),
    (df_laptops["Inches"] > 14),
    (df_laptops["Inches"] > 12),
    (df_laptops["Inches"] <= 12)
]

# If I don't specify the ranges (e.g., greater than 14 and less than or equal to 16),
# it will still work as long as the conditions are ordered correctly.
# In this example, they are ordered from highest to lowest (descending order).
# If I change the order, the result may be incorrect because np.select()
# returns the first matching condition.
# As a good practice, it is better to define explicit ranges:

# conditions = [
#     (df_laptops["Inches"] > 16),
#     (df_laptops["Inches"] > 14) & (df_laptops["Inches"] <= 16),
#     (df_laptops["Inches"] > 12) & (df_laptops["Inches"] <= 14),
#     (df_laptops["Inches"] <= 12)
# ]

values = ["Too Big", "Big", "Small", "Too Small"]
```

```python
df_laptops["Screen_size"] = np.select(conditions, values)
```

```python
df_laptops.head()
```

```python
df_laptops["Screen_size"].value_counts()
```
