### 7. Selecting with conditions

### 16. query()

```python
df = pd.read_csv("players_20.csv")
df = df.set_index("short_name")
columns = ["long_name", "age", "dob", "height_cm", "weight_kg", "nationality", "club"]
df = df[columns]
df.head()
```

```python
df.query("age > 34")
```

```python
df[df["age"] > 34]
```

```python
df.query("age > 34 and nationality == 'Brazil'")
```

```python
filter = (df["age"] > 34) & (df["nationality"] == "Brazil")
df[filter]
```

```python
df.query("not (age > 34)")
```

```python
df[~(df["age"] > 34)]
```

```python
df.query("height_cm / 100 > 1.8")
```

```python
df[df["height_cm"]/100 > 1.8]
```

```python
df.dtypes
```

```python
df["dob"] = df["dob"].astype("datetime64[s]")
```

```python
df["dob"].dt.year
```

```python
df.query("dob.dt.year > 2000")
```

```python
df[df["dob"].dt.year > 2000]
```

### 17. apply()

```python
df["age"].apply(np.sqrt)
```

```python
def calculate_bmi(row):
    row["height_cm"] = row["height_cm"] / 100
    bmi = row["weight_kg"] / row["height_cm"]**2
    
    return bmi
```

```python
df.apply(calculate_bmi, axis=1)
```

### 20. copy()

#### 20.1. deep = True

by defalt -> moditications to the data and indices of the df are not reflected in the copy

```python
df_copy = df.copy()
df_copy.head()
```

```python
df.loc["L. Messi","height_cm"] = 180
```

```python
df.loc["L. Messi", "height_cm"]
```

```python
df_copy.loc["L. Messi", "height_cm"]
```

#### 20.2. deep = False

by defalt -> moditications to the data and indices of the df are reflected in the copy

```python
df_copy = df.copy(deep=False)
df_copy.head()
```

```python
df.loc["L. Messi","weight_kg"] = 62
```

```python
df.loc["L. Messi", "weight_kg"]
```

```python
df_copy.loc["L. Messi", "weight_kg"]
```
