### 1. Configuration

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("players_20.csv")
df.head()
```

### 2. Core Concepts

loc()  -> we have to specify rows and columns based on their index/row and columns labels. Write the name of the index or columns to get the data.

iloc() -> we have to specify rows and columns by their inter positions values. 

### 3. Configing the df

```python
df = df.set_index("short_name")
```

```python
df.head()
```

```python
columns = ["long_name", "age", "dob", "height_cm", "weight_kg", "nationality", "club"]
df = df[columns]
df.head()
```

### 4. Selecting with a single value

loc[row_label, column_label]

```python
df.loc["L. Messi"]
```

```python
df.loc["L. Messi", "height_cm"]
```

```python
df.loc["Cristiano Ronaldo", "weight_kg"]
```

```python
df.loc[:,"height_cm"]
```

```python
df.loc["L. Messi", :] # the : is optional in this case
```

### 5. Selecting with a multiple values

```python
players = ["L. Messi", "Cristiano Ronaldo"]
df.loc[players]
```

```python
df.loc[players, "height_cm"]
```

```python
columns = ["height_cm", "weight_kg"]
df.loc["L. Messi", columns]
```

```python
df.loc[players, columns]
```

### 6. Selecting a range of data with a slice

start:stop:step -> both, start and stop elements are included

```python
players = ["L. Messi", "Cristiano Ronaldo"]

df.loc[players, "age":"club"]
```

```python
df.index[:10]
```

```python
df.loc["L. Messi":"M. Salah", "age":"weight_kg"]
```

### 7. Selecting with conditions

```python
columns = ["age", "dob", "height_cm", "weight_kg"]

# players with height above 180cm

height = 180
df.loc[df["height_cm"] > height, columns]
```

```python
columns = ["age", "dob", "height_cm", "weight_kg"]

# players with height above 180cm and from Brazil

height = 180
nacionality = "Brazil"

df.loc[(df["height_cm"] > height) & (df["nationality"] == nacionality), columns]
```
