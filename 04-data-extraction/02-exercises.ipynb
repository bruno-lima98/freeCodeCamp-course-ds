### 1. Configuration

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("players_20.csv")
df = df.set_index("short_name")
df.head()
```

### 2. Selecting with a single value

iloc[row_label, column_label]

```python
columns = ["long_name", "age", "dob", "height_cm", "weight_kg", "nationality", "club"]
df = df[columns]
```

```python
df.iloc[0]["height_cm"]
```

```python
df.iloc[0,3]
```

```python
df.iloc[1]["weight_kg"]
```

```python
df.iloc[:]["height_cm"]
```

```python
df.iloc[0, :]
```

### 3. Selecting with a multiple values

```python
players_index = [0, 1]
df.iloc[players_index]
```

```python
df.iloc[players_index]["height_cm"]
```

```python
columns = [3,4]
df.iloc[0, columns]
```

```python
df.iloc[0][["height_cm", "weight_kg"]]
```

```python
df.iloc[players_index][["height_cm", "weight_kg"]]
```

```python
df.iloc[players_index, columns]
```

### 4. Selecting a range of data with a slice

start:stop:step -> both, start and stop elements are included

```python
players_index = [0, 1]

df.iloc[players_index, 1:]
```

```python
columns = [1, 2, 3, 4]

df.iloc[0:10, columns]
```

```python
df.iloc[0:10, 1:5]
```

### 5. Selecting with conditions

iloc cannot accept a boolean Series, but only a boolean list. We have to use the list() function to convert a Series into a boolean list

```python
columns = [1, 2, 3, 4]
height = 180

df.iloc[list(df["height_cm"] > 180), columns]
```

```python
columns = [1, 2, 3, 4]
height = 180
nacionality = "Brazil"

df.iloc[list((df["height_cm"] > height) & (df["nationality"] == nacionality)), columns]
```

```python
columns = [1, 2, 3, 4]
height = 180
nacionality = "Brazil"

filter = list(
    (df["height_cm"] > height) & 
    (df["nationality"] == nacionality)
    )

df.iloc[filter, columns]
```
