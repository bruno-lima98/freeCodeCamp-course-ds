### 1. Configuration

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Car_sales.csv")
df.head()
```

```python
columns = [
    "Manufacturer", "Sales_in_thousands", "Vehicle_type", 
    "Price_in_thousands", "Engine_size", "Horsepower", 
    "Fuel_capacity"
    ]

df = df[columns]
df.head()
```

```python
df.nunique()
```

### 2. agg()

```python
numerical = []
string = []

for i in columns:
    if df.dtypes[i] == "O":
        string.append(i)
    else:
        numerical.append(i)
```

```python
df[numerical].agg("sum")
```

```python
df[numerical].agg("mean")
```

```python
df[numerical].agg(["mean", "count", "sum"])
```

```python
df[numerical].agg({
    "Sales_in_thousands":["sum", "mean"],
    "Price_in_thousands":["sum", "max"]
    }
)
```

```python
df[["Sales_in_thousands", "Price_in_thousands"]].agg("sum", axis=1)
```

```python
df.agg(sum_of_sales = ("Sales_in_thousands", "sum"),
       sum_of_price = ("Price_in_thousands", "sum"))
```

### 3. The split-apply-combine strategy

```python
df["Vehicle_type"].value_counts()
```

```python
passanger_filter = df["Vehicle_type"] == "Passenger"
car_filter       = df["Vehicle_type"] == "Car"
```

```python
passanger_mean = df[passanger_filter]["Sales_in_thousands"].mean()
```

```python
car_mean = df[car_filter]["Sales_in_thousands"].mean()
```

```python
pd.DataFrame({
    "Vehicle_type":["Car", "Passenger"],
    "Sales_in_thousands":[car_mean, passanger_mean]
    })
```
