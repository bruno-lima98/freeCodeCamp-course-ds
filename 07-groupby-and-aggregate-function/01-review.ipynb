### 1. Configuration

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Car_sales.csv")
df.head()
```

```python
columns = [
    "Manufacturer", "Sales_in_thousands", "Vehicle_type", 
    "Price_in_thousands", "Engine_size", "Horsepower", 
    "Fuel_capacity"
    ]

df = df[columns]
df.head()
```

```python
df.nunique()
```

### 2. agg()

```python
numerical = []
string = []

for i in columns:
    if df.dtypes[i] == "O":
        string.append(i)
    else:
        numerical.append(i)
```

```python
df[numerical].agg("sum")
```

```python
df[numerical].agg("mean")
```

```python
df[numerical].agg(["mean", "count", "sum"])
```

```python
df[numerical].agg({
    "Sales_in_thousands":["sum", "mean"],
    "Price_in_thousands":["sum", "max"]
    }
)
```

```python
df[["Sales_in_thousands", "Price_in_thousands"]].agg("sum", axis=1)
```

```python
df.agg(sum_of_sales = ("Sales_in_thousands", "sum"),
       sum_of_price = ("Price_in_thousands", "sum"))
```

### 5. groupby() and agg()

```python
df.groupby("Vehicle_type")[numerical].agg(["min", "max", "mean"]).round(1)
```

```python
df.groupby("Vehicle_type")[numerical].agg(
                                        min_engine=("Engine_size", "min"),
                                        max_horse=("Horsepower", "max")
                                        )
```

```python
df.groupby("Manufacturer")[numerical].agg(
                                        sum_sales = ("Sales_in_thousands", "sum"),
                                        mean_price = ("Price_in_thousands", "mean")
                                        )
```

### 7. filter()

```python
def filter_func(x):
    return x["Sales_in_thousands"].sum() > 52

df.groupby("Manufacturer")[numerical].filter(filter_func)
```

```python
df.groupby("Manufacturer")["Sales_in_thousands"].sum().sort_values(ascending=True)
```

```python
df[df["Manufacturer"].isin(["Porsche", "Jaguar", "Saab", "Infiniti", "Audi", "BMW"])]
```

```python
# agg() → reduz cada grupo a um resumo (uma linha por grupo).
# transform() → calcula algo por grupo, mas devolve o mesmo número de linhas do DataFrame original.
# filter() → decide se o grupo inteiro fica ou sai, preservando todas as linhas dos grupos aprovados.
```
